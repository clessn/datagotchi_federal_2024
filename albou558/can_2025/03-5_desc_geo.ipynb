```r
# On récupère la carte des circonscriptions de 2022
map_data <- cartessn::spatial_canada_2022_electoral_ridings

# On repasse en projection lat/lon
map_data <- st_transform(map_data, crs = 4326)
```

```r
p_main <- ggplot(map_data) +
  geom_sf(aes(fill = my_variable), color = NA) +
  # Choisissez la palette de couleurs qui vous convient :
  scale_fill_viridis_d(option = "C", na.value = "grey80") +
  theme_minimal() +
  labs(
    title = "Carte globale des circonscriptions canadiennes",
    fill = "Ma variable"  # Légende
  )
```

```r
# Québec
quebec_data <- cartessn::crop_map(map_data, "quebec_city")
p_quebec <- ggplot(quebec_data) +
  geom_sf(aes(fill = my_variable), color = NA) +
  scale_fill_viridis_d(option = "C", na.value = "grey80") +
  theme_void() +
  labs(title = "Québec")

# Montréal (souvent "montreal" ou "montreal_region")
montreal_data <- cartessn::crop_map(map_data, "montreal")
p_montreal <- ggplot(montreal_data) +
  geom_sf(aes(fill = my_variable), color = NA) +
  scale_fill_viridis_d(option = "C", na.value = "grey80") +
  theme_void() +
  labs(title = "Montréal")

# Vancouver
vancouver_data <- cartessn::crop_map(map_data, "vancouver")
p_vancouver <- ggplot(vancouver_data) +
  geom_sf(aes(fill = my_variable), color = NA) +
  scale_fill_viridis_d(option = "C", na.value = "grey80") +
  theme_void() +
  labs(title = "Vancouver")


# Winnipeg
winnipeg_data <- cartessn::crop_map(map_data, "winnipeg")
p_winnipeg <- ggplot(winnipeg_data) +
  geom_sf(aes(fill = my_variable), color = NA) +
  scale_fill_viridis_d(option = "C", na.value = "grey80") +
  theme_void() +
  labs(title = "Winnipeg")

# Toronto (souvent "toronto" ou "gta" ou "toronto_gta")
gta_data <- cartessn::crop_map(map_data, "toronto")
p_gta <- ggplot(gta_data) +
  geom_sf(aes(fill = my_variable), color = NA) +
  scale_fill_viridis_d(option = "C", na.value = "grey80") +
  theme_void() +
  labs(title = "GTA / Toronto")
```

```r
# On combine le tout.
# Ici, on fait une disposition verticale (ncol = 1) :
p_final <- p_main + 
  (p_quebec + p_montreal + p_vancouver) +
  (p_winnipeg + p_gta) +
  plot_layout(ncol = 1)

# On affiche le résultat
p_final
```
