```r
circo_prov <- read.csv("../../_SharedFolder_datagotchi_federal_2024/data/clustering/qc_2022/postCircProv.csv", sep = ";", encoding = "latin1")
```

```r
circo_prov
```

```r
library(dplyr)
library(stringr)
library(tidyr)

# 1. Préparation de la table des circonscriptions (circo_prov)
# Extraction des 3 premiers caractères du code postal (CO_POSTL)
circo_prov <- circo_prov %>%
  mutate(postal_code = str_sub(CO_POSTL, 1, 3))

# Création d'une table de correspondance par code postal (3 caractères)
mapping <- circo_prov %>%
  group_by(postal_code) %>%
  summarize(
    circonscriptions = list(unique(NM_CIRC)),
    nb_circos = n_distinct(NM_CIRC)
  )

# 2. Affichage du mapping complet en "aplatissant" la liste
mapping_complete <- mapping %>%
  unnest(circonscriptions)

cat("Mapping complet :\n")
print(mapping_complete)

```

```r
library(dplyr)

# Si nécessaire, convertir les colonnes de type liste en chaînes de caractères :
mapping <- mapping %>% 
  mutate_if(is.list, ~ sapply(., toString))

write.csv(mapping, file = "../../_SharedFolder_datagotchi_federal_2024/data/clustering/qc_2022/mapping.csv", row.names = FALSE)
```

```r
library(dplyr)
library(stringr)
library(tidyr)

# --- Partie concernant les répondants ---

# 2. Joindre le mapping sur le préfixe postal
datagotchi_joined <- datagotchi %>%
  left_join(mapping, by =  "postal_code")
```

```r
datagotchi_joined
```

```r

# 3. Filtrer les répondants dont le nombre de circonscriptions possibles est supérieur à 1
repondants_ambigus <- datagotchi_joined %>%
  filter(nb_circos.y > 1)
```

```r
# 4. Compter le nombre de répondants ambigus
nb_ambigus <- nrow(repondants_ambigus)
cat("Nombre de répondants dont la circonscription est incertaine (nb_circos > 1) :", nb_ambigus, "\n")
```
