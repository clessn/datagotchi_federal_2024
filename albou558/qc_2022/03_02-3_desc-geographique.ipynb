```r
library(dplyr)
library(stringr)
library(tidyr)

# --- Création du mapping à partir de circo_prov ---
# Extraction des 3 premiers caractères du code postal (CO_POSTL)
circo_prov <- circo_prov %>%
  mutate(postal_code = str_sub(CO_POSTL, 1, 3))

# Création du mapping : pour chaque préfixe, liste des circonscriptions et nombre distinct
mapping <- circo_prov %>%
  group_by(postal_code) %>%
  summarize(
    circonscriptions = list(unique(NM_CIRC)),
    nb_circos = n_distinct(NM_CIRC)
  )

# (Optionnel) Affichage du mapping complet en "aplatissant" la liste
mapping_complete <- mapping %>% unnest(circonscriptions)
cat("Mapping complet :\n")
print(mapping_complete)

# --- Association aux répondants dans 'datagotchi' ---

# Extraction des 3 premiers caractères du code postal des répondants
datagotchi <- datagotchi %>%
  mutate(postal_prefix = str_sub(postal_code, 1, 3))

# Jointure avec le mapping sur le préfixe postal
datagotchi_joined <- datagotchi %>%
  left_join(mapping, by = c("postal_prefix" = "postal_code"))

# --- Filtrer les répondants non ambigus ---
# On garde uniquement les répondants dont le préfixe correspond à une seule circonscription (nb_circos == 1)
datagotchi_non_ambigu <- datagotchi_joined %>%
  filter(nb_circos == 1)

# Affichage du nombre de répondants non ambigus
cat("Nombre de répondants dont la circonscription n'est pas ambiguë :", nrow(datagotchi_non_ambigu), "\n")
```

```r
library(dplyr)
library(tidyr)
library(ggplot2)
library(sf)
library(patchwork)

# --- 0. Convertir le shapefile en polygones individuels ---
shp_single <- st_cast(shp, "POLYGON")

# --- 1. Agréger les données de clusters par circonscription ---
# On suppose que dans datagotchi_non_ambigu la variable de circonscription s'appelle "circonscriptions"
# et que les colonnes pour les clusters s'appellent cluster_1, cluster_2, ..., cluster_6.
agg_clusters <- datagotchi_non_ambigu %>%
  mutate(circonscriptions = as.character(circonscriptions)) %>%
  group_by(circonscriptions) %>%
  summarize(
    cluster_1 = mean(cluster_1, na.rm = TRUE),
    cluster_2 = mean(cluster_2, na.rm = TRUE),
    cluster_3 = mean(cluster_3, na.rm = TRUE),
    cluster_4 = mean(cluster_4, na.rm = TRUE),
    cluster_5 = mean(cluster_5, na.rm = TRUE),
    cluster_6 = mean(cluster_6, na.rm = TRUE)
  )
```

```r
# --- 2. Créer une data frame par cluster ---
# Chaque data frame contiendra 2 colonnes :
# - "circonscription" (la circonscription, issue de la variable "circonscriptions")
# - "percentage" (le pourcentage de répondants dans le cluster)
list_cluster_dfs <- lapply(1:6, function(i) {
  cluster_col <- paste0("cluster_", i)
  agg_clusters %>%
    select(circonscription = circonscriptions, percentage = !!sym(cluster_col))
})
names(list_cluster_dfs) <- paste0("cluster_", 1:6)
```

```r

# --- 3. Joindre ces agrégations au shapefile ---
# Dans votre shapefile 'shp', la variable identifiant la circonscription est supposée être NM_CEP.
# Pour chaque cluster, on joint la data frame correspondante.
list_shp_cluster <- lapply(1:6, function(i) {
  cluster_name <- paste0("cluster_", i)
  shp_single %>%
    left_join(list_cluster_dfs[[cluster_name]],
              by = c("NM_CEP" = "circonscription"))
})
names(list_shp_cluster) <- paste0("cluster_", 1:6)
```

```r
# --- 4. Définir une couleur haute spécifique par cluster ---
cluster_colors <- c(
  "cluster_1" = "blue",
  "cluster_2" = "red",
  "cluster_3" = "green",
  "cluster_4" = "purple",
  "cluster_5" = "orange",
  "cluster_6" = "brown"
)

# --- 5. Définir des titres personnalisés pour chaque cluster ---
cluster_titles <- c(
  "cluster_1" = "Julie",
  "cluster_2" = "Jean-Guy",
  "cluster_3" = "Mélanie",
  "cluster_4" = "Karim",
  "cluster_5" = "Charlie",
  "cluster_6" = "Jacques"
)

# --- 6. Créer un graphique pour chaque cluster sans afficher les noms des circonscriptions ---
plots <- lapply(1:6, function(i) {
  cluster_name <- paste0("cluster_", i)
  df <- list_shp_cluster[[cluster_name]]
  ggplot() +
    geom_sf(data = df, aes(fill = percentage), color = "grey50", size = 0.3) +
    scale_fill_gradient(
      low = "darkgrey",
      high = cluster_colors[cluster_name],
      name = "Pourcentage"
    ) +
    ggtitle(paste("Distribution", cluster_titles[cluster_name])) +
    theme_minimal() +
    theme(plot.title = element_text(hjust = 0.5))
})

# --- 7. Afficher chacun des graphiques individuellement ---
for (p in plots) {
  print(p)
}
```
