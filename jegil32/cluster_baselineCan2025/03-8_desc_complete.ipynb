```r
# Charger les packages nécessaires
library(nnet)
library(dplyr)

# S'assurer que la variable cluster est un facteur
DataValidation$cluster_name <- as.factor(DataValidation$cluster_name)
DataValidation <- droplevels(DataValidation)

# Définir les variables explicatives à partir de la liste (les noms de variables de vars_to_print)
predictor_vars <- names(vars_to_print)

# Construction de la formule : cluster ~ var1 + var2 + ... + varN
formula_str <- paste("cluster_name ~", paste(predictor_vars, collapse = " + "))
model_formula <- as.formula(formula_str)

# Ajustement du modèle multinomial
multinom_model <- multinom(model_formula, data = DataValidation, maxNWts = 5000)

# Affichage du résumé du modèle
summary(multinom_model)
```

```r
# Calcul des probabilités prédites pour chaque observation
predicted_probs <- predict(multinom_model, newdata = DataValidation, type = "probs")

# Affichage des premières lignes des probabilités prédites
head(predicted_probs)
```
